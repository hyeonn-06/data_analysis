```python
# 딕셔너리 데이터 작성 (json 형식과 같다.)
d = {"name":"혼자 공부하는 데이터 분석"}
print(d)
print(type(d))
```

```python
# 딕셔너리를 JSON 문자열로 변환 (json.dumps())
import json
d_str = json.dumps(d, ensure_ascii=False)
print(d_str)
print(type(d_str))
```

```python
# JSON 문자열을 딕셔너리로 변환 (json.loads())
d2 = json.loads(d_str)
print(d2)
print(type(d2))
```

```python
# JSON 리스트를 딕셔너리 리스트로 변환
d4_str= """
[
  {"name": "혼자 공부하는 데이터 분석", "author": "박해선", "year":2022},
  {"name": "혼자 공부하는 머신러닝+딥러닝", "author": "박해선", "year":2020}
]
"""
d4 = json.loads(d4_str)
print(d4)
print(type(d4))
```

```python
# JSON 문자열을 데이터프레임으로 변환1
# Pandas 2.1 버전부터는 read_json(), read_xml() 함수에 파일 경로나 파일 객체를 전달해야 함
# 원시 문자열은 오류 발생 -> StringIO 클래스로 감싸면 파일 객체처럼 다룰 수 있음
from io import StringIO
import pandas as pd
pd.read_json(StringIO(d4_str))
```

```python
# JSON 문자열을 데이터프레임으로 변환2
# JSON 문자열을 파이썬 객체로 만든 다음 DataFrame 클래스 사용
pd.DataFrame(d4)
```

```python
# XML 문자열을 파이썬 객체로 변환
x_str = """
<book>
  <name>혼자 공부하는 데이터 분석</name>
  <author>박해선</author>
  <year>2022</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et
book = et.fromstring(x_str)
print(type(book))
```

```python
print(book.tag)
```

```python
book_childs = list(book)
print(book_childs)
```

```python
name, author, year = book_childs
print(name.text)
print(author.text)
print(year.text)
```

```python
name = book.findtext("name")
author = book.findtext("author")
year = book.findtext("year")
print(name)
print(author)
print(year)
```

```python
x2_str="""
<books>
  <book>
    <name>혼자 공부하는 데이터 분석</name>
    <author>박해선</author>
    <year>2022</year>
  </book>
  <book>
    <name>혼자 공부하는 머신러닝+딥러닝</name>
    <author>박해선</author>
    <year>2020</year>
  </book>
</books>
"""
```

```python
books = et.fromstring(x2_str)
print(books.tag)
```

```python
for book in books.findall("book"):
  name = book.findtext("name")
  author = book.findtext("author")
  year = book.findtext("year")
  print(name, author, year, sep=', ')
```
