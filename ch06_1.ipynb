```python
import gdown
gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)

import pandas as pd
ns_book7 = pd.read_csv("ns_book7.csv", low_memory=False)
ns_book7.head()
```

```python
# 상위 30개 출판사 목록 구하기
# value_counts() : 등장 횟수(count)가 많은 순으로 결과를 내림차순 정렬
top30_pubs = ns_book7["출판사"].value_counts()[:30]
top30_pubs
```

```python
# isin() :  특정 값들이 포함되어 있는지를 확인 후 True / False 반환
top30_pubs_idx = ns_book7["출판사"].isin(top30_pubs.index) # 상위 30개에 해당하는 출판사는 True / 그렇지 않으면 False
top30_pubs_idx
```

```python
# 상위 30개 출판사의 발행 도서 개수(count)의 총합 구하기
top30_pubs_idx.sum()
```

```python
# 상위 30개 출판사의 발행도서 51886개 중 1000개만 임의로 선택
ns_book8 = ns_book7[top30_pubs_idx].sample(1000, random_state=42)
ns_book8.head()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(ns_book8["발행년도"], ns_book8["출판사"])
ax.set_title("출판사별 발행 도서")
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
# scatter() 함수의 마커 사이즈 변경
# s 매개변수 : 마커의 사이즈 지정 (default 값 = 36(rcParams["lines.markersize"]의 기본 값은 6의 제곱))
ax.scatter(ns_book8["발행년도"], ns_book8["출판사"], s=ns_book8["대출건수"])
ax.set_title("출판사별 발행 도서")
fig.show()
```

```python
fig, ax = plt.subplots(figsize=(10,8))
# 마커 꾸미기
# linewidth : 마커 테두리 선의 두께 지정 (default : 1.5)
# edgecolors : 마커 테두리의 색 지정 (default : "face"(마커의 색을 의미))
# alpha : 마커의 투명도 지정
# c : 산점도의 색 지정
ax.scatter(ns_book8["발행년도"], ns_book8["출판사"], s=ns_book8["대출건수"]*2, linewidths=0.5, edgecolors="k", alpha=0.3, c=ns_book8["대출건수"])
ax.set_title("출판사별 발행 도서")
fig.show()
```

```python
# 1. 도화지(fig)와 그래프 영역(ax)를 생성, 크기는 10x8인치로 설정
fig, ax = plt.subplots(figsize=(10, 8))
# 2. 산점도(scatter plot) 생성
sc = ax.scatter(
    ns_book8["발행년도"],            # X축: 발행년도
    ns_book8["출판사"],             # Y축: 출판사 (카테고리별로 배치됨)
    s=ns_book8["대출건수"]**1.3,    # 마커 크기: 대출건수를 1.3제곱해서 강조 (값이 클수록 더 큼)
    linewidths=0.5,                # 마커 테두리 두께
    edgecolors="k",               # 테두리 색상: 검정("k"는 black)
    alpha=0.3,                    # 마커 투명도 (0.3이면 많이 투명함)
    c=ns_book8["대출건수"],         # 마커 색상: 대출건수에 따라 색깔 다르게 표시
    cmap="jet"                    # 색상 맵: jet 컬러맵 (무지개 느낌)
)
# 3. 그래프 제목 설정
ax.set_title("출판사별 발행 도서")
# 4. 산점도 컬러맵에 대응하는 컬러바 추가 (값의 범위에 따른 색상 설명)
fig.colorbar(sc)
# 5. 최종적으로 그래프를 출력 (Jupyter/Colab 외부 환경에서 사용하는 경우에 유용)
fig.show()
```
