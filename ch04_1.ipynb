```python
import gdown
gdown.download("https://bit.ly/3736JW1","ns_book6.csv",quiet=False)
```

```python
import pandas as pd
ns_book6 = pd.read_csv("ns_book6.csv", low_memory=False)
ns_book6.head()
```

```python
# describe() : 수치형 열에 대한 요약 통계를 보여줌
# count : 누락된 값을 제외한 데이터 개수 / mean : 평균 / std : 표준편차 / min : 최솟값 / 25%,75% : 25% 지점,75% 지점에 놓인 값 / 50% : 중앙값 / max : 최댓값
ns_book6.describe()
```

```python
sum(ns_book6['도서권수'] == 0)
```

```python
ns_book7 = ns_book6[ns_book6['도서권수'] > 0]
ns_book7.head()
```

```python
ns_book7.describe()
```

```python
# % 0.25 / 0.5 / 0.75가 기본 -> percentiles = []로 변경 가능
ns_book7.describe(percentiles=[0.3, 0.6, 0.9])
```

```python
# 다른 데이터 타입 열의 기술 통계가 보고 싶으면 include=?
# count : 누락된 값을 제외한 데이터 개수 / unique : 고유한 데이터 개수 / top : 가장 많이 등장하는 데이터 / freq : 가장 많이 등장한 빈도
ns_book7.describe(include='object')
```

```python
# 평균 구하기
x = [10,20,30]
sum = 0
for i in range(len(x)):
  sum += x[i]
print(f"평균: {sum/len(x)}")
```

```python
# 평균 구하기 : mean()
print(ns_book7['대출건수'].mean())
```

```python
# 중앙값 구하기 : median()
print(ns_book7['대출건수'].median())
# 데이터 개수가 짝수면 가운데 2개 값의 평균을 중앙값으로 출력
temp_df = pd.DataFrame([1,2,3,4])
print(temp_df.median())
```

```python
# 중복 제거하고 중앙값 구하기
ns_book7['대출건수'].drop_duplicates().median()
```

```python
# 최솟값 구하기 : min()
ns_book7['대출건수'].min()
```

```python
# 최댓값 구하기 : max()
ns_book7['대출건수'].max()
```

```python
# 분위수(데이터를 순서대로 늘어 놓았을 때 이를 균등한 간격으로 나누는 기준점) 구하기 : quantile() (0(0%) ~ 1(100%)까지의 숫자)
# 1개의 분위수 지정
print(ns_book7['대출건수'].quantile(0.25))
# 1개 이상의 분위수 지정(리스트 활용)
print(ns_book7['대출건수'].quantile([0, 0.25, 0.5, 0.75, 1]))
```

```python
# 분산 구하기 : var()
ns_book7['대출건수'].var()
```

```python
# 표준편차 구하기 : std()
ns_book7['대출건수'].std()
```

```python
# 최빈값 구하기 : mode()
print(ns_book7['도서명'].mode())
print(ns_book7['발행년도'].mode())
```

```python
# 데이터프레임에서 기술 통계 구하기
# numeric_only : 수치형 열만 연산할지를 결정하는 매개변수
ns_book7.mean(numeric_only=True)
```

```python
ns_book7.loc[:,'도서명':].mode() # 모든 행을 비교해 구한 '도서명' ~ 끝 열까지의 각 열의 최빈값
```

```python
ns_book7.to_csv('ns_book7.csv', index=False)
```
