```python
ns_book4.replace({'발행년도': {'2021': 21}})[100:102] # replace() : 새로운 데이터프레임을 반환하므로 대괄호([])를 사용한 인덱싱 가능
```

```python
# 숫자 찾기
# \d\d\d\d : 4자리 숫자
# \d\d(\d\d) : 마지막 2자리를 한 그룹으로 묶음 (그룹은 앞부터 차례대로 \1, \2)
# \d\d(\d\d) : '\1' -> 4자리 숫자에서 마지막 2자리만 그룹으로 묶어 추출
ns_book4.replace({'발행년도': {r'\d\d(\d\d)': r'\1'}}, regex=True)[100:102]
```

```python
# \d\d\d\d = \d{2}\d{2}
ns_book4.replace({'발행년도': {r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]
```

```python
# (.*) : '.'은 모든 문자 대응함 / '*'은 반복 횟수가 0번 이상이라는 의미
# \s : 공백 문자
# \(지은이\) : 그룹으로 지정하지 않고 일반 문자로 나타내기 위해서는 괄호 앞에 \ 추가
ns_book4.replace({'저자': {r'(.*)\s\(지은이\)(.*)\s\(옮긴이\)': r'\1\2'}, '발행년도': {r'\d{2}(\d{2})': r'\1'}}, regex=True)[100:102]
```

```python
# 잘못된 값 수정
# 오류 발생 : '1988.' 숫자가 아닌 문자가 포함되어있음
ns_book4.astype({'발행년도': 'int64'})
```

```python
# contains() : 주어진 문자열이 포함된 모든 문자열을 찾음
ns_book4['발행년도'].str.contains('1988').sum()
```

```python
# \D : 숫자가 아닌 모든 문자
# na=True : 누락된 행을 True로 표시 (누락된 행도 처리하기 위함)
invalid_number = ns_book4['발행년도'].str.contains('\D', na=True)
print(invalid_number.sum())
ns_book4[invalid_number].head()
```

```python
ns_book5 = ns_book4.replace({r'발행년도': r'.*(\d{4}).*'}, r'\1', regex=True)
ns_book5[invalid_number].head()
```

```python
unknown_year = ns_book5['발행년도'].str.contains('\D', na=True)
print(unknown_year.sum()) # 변환되지 않은 값(NaN or 4자리 숫자가 아닌 값)의 수
ns_book5[unknown_year].head()
```

```python
# 변환되지 않은 값들은 -1으로 변경
ns_book5.loc[unknown_year,'발행년도'] = -1
ns_book5 = ns_book5.astype({'발행년도': 'int64'})
```

```python
# 누락된 정보를 채우기 위한 라이브러리
import requests
from bs4 import BeautifulSoup
```

```python
#
def get_book_title(isbn):
  # Yes24 도서 검색 페이지 URL
  url = "http://www.yes24.com/Product/Search?domain=BOOK&query={}"
  # URL에 ISBN을 넣어 HTML을 가져옴
  r = requests.get(url.format(isbn))
  soup = BeautifulSoup(r.text, 'html.parser')
  # 클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져옴
  title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  return title
```

```python
# 정규표현식 라이브러리
import re

def get_book_info(row):
  title = row['도서명']
  author = row['저자']
  pub = row['출판사']
  year = row['발행년도']
  # Yes24 도서 검색 페이지 URL
  url = "http://www.yes24.com/Product/Search?domain=BOOK&query={}"
  # URL에 ISBN을 넣어 HTML을 가져옴
  r = requests.get(url.format(row['ISBN']))
  soup = BeautifulSoup(r.text, 'html.parser')

  try:
    if pd.isna(title):
     # 클래스 이름이 'gd_name'인 <a> 태그의 텍스트를 가져옴
     title = soup.find('a', attrs={'class':'gd_name'}).get_text()
  except AttributeError:
    pass

  try:
    if pd.isna(author):
      # 클래스 이름이 'info_auth'인 <span> 태그 아래 <a> 태그의 텍스트를 가져옴
      authors = soup.find('span', attrs={'class':'info_auth'}).find_all('a')
      author_list = [auth.get_text() for auth in authors]
      author = ', '.join(author_list)
  except AttributeError:
    pass

  try:
    if pd.isna(pub):
      # 클래스 이름이 'info_pub'인 <span> 태그 아래 <a> 태그의 텍스트를 가져옴
      pub = soup.find('span', attrs={'class':'info_pub'}).find('a').get_text()
  except AttributeError:
    pass

  try:
    if year == -1:
      # 클래스 이름이 'info_date'인 <span> 태그 아래 텍스트를 가져옴
      year_str = soup.find('span', attrs={'class':'info_date'}).get_text()
      # 정규 표현식으로 찾은 값 중에 첫번째 것만 사용
      year = re.findall(r'\d{4}', year_str)[0]
  except AttributeError:
    pass
  return title, author, pub, year
```
