```python
# 파일 다운로드 라이브러리
import gdown
# 파일 다운로드
gdown.download("https://bit.ly/3RhoNho",'ns_202104.csv',quiet=False)
```

```python
# 데이터 분석 라이브러리
import pandas as pd
# csv 파일을 데이터 프레임으로 변환
ns_df = pd.read_csv('ns_202104.csv', low_memory=False)
# 데이터 프레임의 첫 5개 행 출력
ns_df.head()
```

```python
# loc() : 데이터프레임의 행과 열을 통해 데이터 선택 메서드
ns_book = ns_df.loc[:,'번호':'등록일자']
# 선택된 데이터 프레임의 첫 5개 행 출력
ns_book.head()
```

```python
# 데이터 프레임의 열 이름 출력
print(ns_df.columns)
```

```python
# index를 통한 열 이름 출력
print(ns_df.columns[0])
```

```python
# 열 이름 비교
ns_df.columns != 'Unnamed: 13'
```

```python
# 조건문을 통한 열 선택
selected_columns = ns_df.columns != 'Unnamed: 13'
# 선택된 열을 통한 데이터 선택
ns_book = ns_df.loc[:,selected_columns]
# 선택된 데이터 프레임의 첫 5개 행 출력
ns_book.head()
```

```python
# 조건문을 통한 열 선택
selected_columns = ns_df.columns != '부가기호'
# 선택된 열을 통한 데이터 선택
ns_book = ns_df.loc[:,selected_columns]
# 선택된 데이터 프레임의 첫 5개 행 출력
ns_book.head()
```

```python
# drop() : 데이터 프레임의 행 또는 열을 삭제하는 메서드
# axis = 0 : 행 삭제(default) / axis = 1 : 열 삭제
# ex) dataframe.drop('삭제할 행 또는 열 이름', axis = ?)
ns_book = ns_df.drop('Unnamed: 13', axis=1)
# 한번에 여러개의 행 또는 열을 삭제하기 위해서는 리스트 사용
ns_book = ns_df.drop(['부가기호','Unnamed: 13'], axis=1)
ns_book.head()
```
